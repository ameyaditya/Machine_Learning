```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#there are two ways of creating a matpllotlib plot
#one is the fucntion method and the object oriented method.
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x** 2
```

```python
x
```

```python
y
```

```python
#Functionsl plot
plt.plot(x,y)
```

```python
#using plt.show() will print the plot, is used outside jupyter notebook
```

```python
#to add xlabel and ylabel
plt.xlabel("X Label")
plt.ylabel("y Label")
plt.title("Test Plot")
plt.plot(x,y)
```

```python
#putting multiple plots on the same canvas
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)

plt.plot(y,x,'b')
```

```python
#object oriented method
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("x Label")
axes.set_ylabel("y label")
axes.set_title("Test")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.45,0.17,0.4,0.3])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title("Outer")
axes2.set_title("Inner")
```

```python
#subplots in Matplotlib
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#the subplots can be accessed using array indexing
for current in axes:
    current.plot(x,y)
```

```python
#the above code creates a canvas with one row and two columns
#fig,axes is a tupple unpacking and wat axes is, is it is just an array
axes
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#the subplots can be accessed using array indexing
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[0].set_title("Left")
axes[1].set_title("Right")

#always make a habbit to put the below code to avoid the plot from overlapping
plt.tight_layout()
```

```python
#using legends
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label = 'Square')
ax.plot(x,x**3, label = 'cube')
ax.legend(loc=0)

#we can save out pots to any formats
fig.savefig('Plottest.png',dpi = 300)
```
