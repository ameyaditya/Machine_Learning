```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x** 2
```

```python
x
```

```python
y
```

```python
#Functionsl plot
plt.plot(x,y)
```

```python
#using plt.show() will print the plot, is used outside jupyter notebook
```

```python
#to add xlabel and ylabel
plt.xlabel("X Label")
plt.ylabel("y Label")
plt.title("Test Plot")
plt.plot(x,y)
```

```python
#putting multiple plots on the same canvas
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)

plt.plot(y,x,'b')
```

```python
#object oriented method
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("x Label")
axes.set_ylabel("y label")
axes.set_title("Test")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.45,0.17,0.4,0.3])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title("Outer")
axes2.set_title("Inner")
```
